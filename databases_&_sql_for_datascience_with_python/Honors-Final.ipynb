```python
import pandas

census = pandas.read_csv('ChicagoCensusData.csv')
crime = pandas.read_csv('ChicagoCrimeData.csv')
public_schools = pandas.read_csv('ChicagoPublicSchools.csv')

census.to_sql("chicago_census_data", con, if_exists='replace', index=False)
crime.to_sql("chicago_crime_data", con, if_exists='replace', index=False)
public_schools.to_sql("chicago_public_schools_data", con, if_exists='replace', index=False)
```

### Exercise 1 Question 1:

```python
%%sql SELECT School.NAME_OF_SCHOOL, Census.COMMUNITY_AREA_NAME, School.AVERAGE_STUDENT_ATTENDANCE
    FROM chicago_public_schools_data AS School
	JOIN chicago_census_data AS Census ON School.COMMUNITY_AREA_NUMBER = Census.COMMUNITY_AREA_NUMBER
    WHERE Census.HARDSHIP_INDEX = 98;
```

### Exercise 1 Question 2:

```python
%%sql SELECT CASE_NUMBER, PRIMARY_TYPE, census.COMMUNITY_AREA_NAME, LOCATION_DESCRIPTION
    FROM chicago_census_data AS census
    RIGHT JOIN chicago_crime_data AS crime ON census.COMMUNITY_AREA_NUMBER = crime.COMMUNITY_AREA_NUMBER
    WHERE crime.LOCATION_DESCRIPTION LIKE '%SCHOOL%';
```

### Exercise 3 Question 3:

```python
%%sql --#SET TERMINATOR @
    CREATE PROCEDURE UPDATE_LEADERS_SCORE
    (IN new_school_id INTEGER, IN new_score INTEGER)
    LANGUAGE SQL
    MODIFIES SQL DATA
    BEGIN
        Update chicago_public_schools_data
        SET Leaders_Score = new_score
        where SCHOOL_ID = new_school_id;

        IF (in_Leaders_Score > 0 AND in_Leaders_Score < 20)
            THEN
            UPDATE chicago_public_schools_data 
            SET Leaders_icon = 'Very weak'
            WHERE School_ID = new_school_id;

        ELSEIF in_Leaders_Score < 40
            THEN
            UPDATE chicago_public_schools_data
            SET Leaders_icon = 'weak'
            WHERE School_ID = new_school_id;

        ELSEIF in_Leaders_Score < 60
            THEN
            UPDATE chicago_public_schools_data
            SET Leaders_icon = 'Average'
            WHERE School_ID = new_school_id;

        ELSEIF in_Leaders_Score < 80
            THEN
            UPDATE chicago_public_schools_data
            SET Leaders_icon = 'Strong'
            WHERE School_ID = new_school_id;

        ELSEIF in_Leaders_Score < 100
            THEN
            UPDATE chicago_public_schools_data
            SET Leaders_icon = 'Very Strong'
            WHERE School_ID = new_school_id;
        END IF;                 
    END
    @
```
